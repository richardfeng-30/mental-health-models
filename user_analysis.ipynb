```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = '/content/drive/My Drive/MindPaw/user_scores_lstm.csv'

df = pd.read_csv(data)

df['mental_health_score_before'] = df['average_positive_probability_before'] * 100
df['mental_health_score_after'] = df['average_positive_probability_after'] * 100

df['score_change'] = df['mental_health_score_after'] - df['mental_health_score_before']

df.head(5)
```

```python
fig, ax = plt.subplots()

ax.hist(df['age'], bins=13, linewidth=0.5, edgecolor="white")

ax.set_title('Age Distribution')
ax.set_xlabel('Age')
ax.set_ylabel('Count')

ax.set(xlim=(8, 20), xticks=np.arange(9, 20),
       ylim=(0, 40), yticks=np.linspace(0, 40, 9))

plt.show()
```

```python
gender_counts = df['gender'].value_counts()

fig, ax = plt.subplots()
ax.bar(gender_counts.index, gender_counts.values, color=['pink', 'blue'])

ax.set_title('Gender Distribution')
ax.set_xlabel('Gender')
ax.set_ylabel('Count')

plt.show()
```

```python
race_counts = df['race_ethnicity'].value_counts()

colors = plt.cm.tab20.colors[:len(race_counts)]

fig, ax = plt.subplots()
bars = ax.bar(race_counts.index, race_counts.values, color=colors)

ax.set_xticks([])

ax.set_title('Race Distribution')
ax.set_xlabel('Race/Ethnicity')
ax.set_ylabel('Count')

ax.legend(bars, race_counts.index, title='Race/Ethnicity')

plt.show()
```

```python
socio_counts = df['socio_economic_status'].value_counts()

fig, ax = plt.subplots()
ax.bar(socio_counts.index, socio_counts.values, color=['yellow', 'red', 'green'])

ax.set_title('Location Distribution')
ax.set_xlabel('Location')
ax.set_ylabel('Count')

plt.show()
```

```python
location_counts = df['urban_rural'].value_counts()

fig, ax = plt.subplots()
ax.bar(location_counts.index, location_counts.values)

ax.set_title('Location Distribution')
ax.set_xlabel('Location')
ax.set_ylabel('Count')

plt.show()
```

```python
parent_counts = df['parental_involvement'].value_counts()

fig, ax = plt.subplots()
ax.bar(parent_counts.index, parent_counts.values)

ax.set_title('Parental Involvement Distribution')
ax.set_xlabel('Parental Involvement')
ax.set_ylabel('Count')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(df['score_change'], bins=20, linewidth=0.5, edgecolor="white")

ax.set_title('Score Change Distribution')
ax.set_xlabel('Score Change')
ax.set_ylabel('Count')

ax.set(xlim=(-80, 80), xticks=np.arange(-70, 100, 10),
       ylim=(0, 60), yticks=np.linspace(0, 60, 9))

plt.show()
```

```python
fig, axs = plt.subplots(3, 2, figsize=(15, 18))

# Histogram for Age Distribution
axs[0, 0].hist(df['age'], bins=13, linewidth=0.5, edgecolor="white")
axs[0, 0].set_title('Age Distribution')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Count')
axs[0, 0].set(xlim=(8, 20), xticks=np.arange(9, 20),
              ylim=(0, 40), yticks=np.linspace(0, 40, 9))

# Bar chart for Gender Distribution
gender_counts = df['gender'].value_counts()
axs[0, 1].bar(gender_counts.index, gender_counts.values, color=['pink', 'blue'])
axs[0, 1].set_title('Gender Distribution')
axs[0, 1].set_xlabel('Gender')
axs[0, 1].set_ylabel('Count')

# Bar chart for Race/Ethnicity Distribution
race_counts = df['race_ethnicity'].value_counts()
colors = plt.cm.tab20.colors[:len(race_counts)]
bars = axs[1, 0].bar(race_counts.index, race_counts.values, color=colors)
axs[1, 0].set_title('Race Distribution')
axs[1, 0].set_xlabel('Race/Ethnicity')
axs[1, 0].set_ylabel('Count')
axs[1, 0].set_xticks([])
axs[1, 0].legend(bars, race_counts.index, title='Race/Ethnicity')

# Bar chart for Socio-economic Status Distribution
socio_counts = df['socio_economic_status'].value_counts()
axs[1, 1].bar(socio_counts.index, socio_counts.values, color=['yellow', 'red', 'green'])
axs[1, 1].set_title('Socio-economic Status Distribution')
axs[1, 1].set_xlabel('Status')
axs[1, 1].set_ylabel('Count')

# Bar chart for Urban/Rural Distribution
location_counts = df['urban_rural'].value_counts()
axs[2, 0].bar(location_counts.index, location_counts.values)
axs[2, 0].set_title('Location Distribution')
axs[2, 0].set_xlabel('Location')
axs[2, 0].set_ylabel('Count')

# Remove the last empty subplot (bottom-right)
fig.delaxes(axs[2, 1])

plt.tight_layout()
plt.show()
```
